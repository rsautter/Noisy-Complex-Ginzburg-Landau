### Libraries

```python
from GPA import GPA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
ga = GPA(0.03)
ga.version()
```

# Without Noise

```python
data = np.load("data/withoutNoise/data.npy")
times = np.load("data/withoutNoise/times.npy")

print(data.shape)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(np.real(data[1000]),cmap='rainbow')
plt.title("Real(Z)",fontsize=15)
plt.subplot(1,3,2)
plt.imshow(np.imag(data[1000]),cmap='rainbow')
plt.title("Imag(Z)",fontsize=15)
plt.subplot(1,3,3)
plt.imshow(np.abs(data[1000]),cmap='rainbow')
plt.title("|Z|",fontsize=15)
plt.show()
```

```python
rz,iz,mz = [],[],[]
for i in tqdm(range(data.shape[0])):
    rz.append(ga(np.real(data[i]),moment=["G1","G2","G3"]))
    iz.append(ga(np.imag(data[i]),moment=["G1","G2","G3"]))
    mz.append(ga(np.abs(data[i]),moment=["G1","G2","G3"]))
rz,iz,mz = pd.DataFrame(rz),pd.DataFrame(iz),pd.DataFrame(mz)
```

```python
rz.to_csv("data/withoutNoise/gpa/real.csv", index=False)
iz.to_csv("data/withoutNoise/gpa/imag.csv", index=False)
mz.to_csv("data/withoutNoise/gpa/mod.csv", index=False)
```

```python
legendSize = 15
titleSize = 18

plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.title("Re(Z)",fontsize=titleSize)
#plt.plot(times, rz["G1"][:-1],color='b')
plt.plot(times, rz["G2"][:-1],color='r',label=r"$G_2$")
plt.plot(times, rz["G3"][:-1],color='k',label=r"$G_3$")
plt.legend(fontsize = legendSize)
plt.xticks([])
plt.xlim(0,max(times))

plt.subplot(3,1,2)
plt.title("Im(Z)",fontsize=titleSize)
#plt.plot(times, iz["G1"][:-1],color='b')
plt.plot(times, iz["G2"][:-1],color='r')
plt.plot(times, iz["G3"][:-1],color='k')
plt.xticks([])
plt.xlim(0,max(times))

plt.subplot(3,1,3)
plt.title("||Z||",fontsize=titleSize)
#plt.plot(times, mz["G1"][:-1],color='b')
plt.plot(times, mz["G2"][:-1],color='r')
plt.plot(times, mz["G3"][:-1],color='k')
plt.xlim(0,max(times))
plt.xlabel("Time")

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,1,1)
plt.plot(twithout, without["G2"][:-1],label="Without Noise", ls= '--')
plt.plot(tadd, add["G2"][:-1],label="Additive Noise", ls= '--')
plt.plot(tmul, mul["G2"][:-1],label="Multiplicative Noise", ls= '--')
plt.plot(tdiff, diff["G2"][:-1],label="Diffusive Noise", ls= '--')
plt.legend()
plt.title("$G_2\Re (Z)$", fontsize=20)
plt.xlim(0,max(twithout))

plt.subplot(2,1,2)
plt.plot(twithout, without["G3"][:-1],label="Without Noise", ls= '--')
plt.plot(tadd, add["G3"][:-1],label="Additive Noise", ls= '--')
plt.plot(tmul, mul["G3"][:-1],label="Multiplicative Noise", ls= '--')
plt.plot(tdiff, diff["G3"][:-1],label="Diffusive Noise", ls= '--')
plt.legend()
plt.title("$G_3(\Re (Z))$", fontsize=20)
plt.xlim(0,max(twithout))

plt.show()
```

```python
labelSize= 15
titleSize= 20
legendSize=12

plt.figure(figsize=(10,15))

plt.subplot(4,1,1)
plt.plot(twithout, without["G2"][:-1],label=r"$G_2(\Re(A))$", ls= '-',color='blue')
plt.plot(twithout, without["G3"][:-1],label=r"$G_3(\Re(A))$", ls= '-',color='red')
plt.axvspan(20,55,alpha=0.15,color='k',label='Spiral formation')
plt.axvspan(55,np.max(twithout),alpha=0.15,color='green',label='Final Stage')


plt.xlim(0,max(twithout))
plt.title("(a) Without Noise",fontsize=titleSize,fontweight="bold")
plt.xlabel("Time",fontsize=labelSize)
plt.legend(fontsize= legendSize)

plt.subplot(4,1,2)
plt.plot(tadd, add["G2"][:-1],label="Additive Noise", ls= '-',color='blue')
plt.plot(tadd, add["G3"][:-1],label="Additive Noise", ls= '-',color='red')
plt.xlim(0,max(tadd))
plt.title("(b) Additive Noise",fontsize=titleSize,fontweight="bold")
plt.xlabel("Time",fontsize=labelSize)

plt.subplot(4,1,3)
plt.plot(tmul, mul["G2"][:-1],label="Multiplicative Noise", ls= '-',color='blue')
plt.plot(tmul, mul["G3"][:-1],label="Multiplicative Noise", ls= '-',color='red')
plt.xlim(0,max(tmul))
plt.xlabel("Time",fontsize=labelSize)
plt.axvspan(15,24,alpha=0.15,color='k',label='Spiral formation')
plt.axvspan(24,np.max(tmul),alpha=0.15,color='green',label='Final Stage')
plt.title("(c) Multiplicative Noise",fontsize=titleSize,fontweight="bold")

plt.subplot(4,1,4)
plt.plot(tdiff, diff["G2"][:-1],label="Diffusive Noise", ls= '-',color='blue')
plt.plot(tdiff, diff["G3"][:-1],label="Diffusive Noise", ls= '-',color='red')
plt.xlim(0,max(tdiff))
plt.title("(d) Diffusive Noise",fontsize=titleSize,fontweight="bold")
plt.axvspan(16,27,alpha=0.15,color='k',label='Spiral formation')
plt.axvspan(27,np.max(tdiff),alpha=0.15,color='green',label='Final Stage')
plt.xlabel("Time",fontsize=labelSize)

plt.tight_layout()
plt.show()
```

# Videos

```python
diff = pd.read_csv("data/diffusiveNoise/gpa/real.csv")
tdiff = np.load("data/diffusiveNoise/times.npy")
data = np.load("data/diffusiveNoise/data.npy")
times = np.load("data/diffusiveNoise/times.npy")
```

```python
from matplotlib import animation
labelSize= 15
titleSize= 20
legendSize=15


def plotFunc(timeStep):
    print(timeStep)
    grid = plt.GridSpec(2, 3)
    
    plt.subplot(grid[0,0])
    plt.imshow(np.real(data[timeStep]),cmap='rainbow')
    plt.title("$\Re(A)$",fontsize=titleSize,fontweight="bold")

    plt.subplot(grid[0,1])
    plt.imshow(np.real(data[timeStep]),cmap='rainbow')
    plt.title("$\Im(A)$",fontsize=titleSize,fontweight="bold")

    plt.subplot(grid[0,2])
    plt.imshow(np.abs(data[timeStep]),cmap='rainbow')
    plt.title("$|A|$",fontsize=titleSize,fontweight="bold")

    plt.subplot(grid[1,:])
    plt.plot(tdiff, diff["G2"][:-1], ls= '-',color='blue',label=r"$G_2(\Re(A))$")
    plt.plot(tdiff, diff["G3"][:-1], ls= '-',color='red', label=r"$G_3(\Re(A))$")
    plt.xlim(0,max(tdiff))
    plt.title("Diffusive Noise",fontsize=titleSize,fontweight="bold")
    plt.axvspan(16,27,alpha=0.15,color='k',label='Spiral formation')
    plt.axvspan(27,np.max(tdiff),alpha=0.15,color='green',label='Final Stage')
    plt.xlabel("Time",fontsize=labelSize)
    plt.axvline(times[timeStep],lw=2,color='k')
    plt.legend()
    
figure = plt.figure(figsize=(15,10))
anim = animation.FuncAnimation(figure, func=plotFunc,frames=np.arange(0,1499), interval=20, repeat=False)
anim.save('DiffusiveGPA.mp4',dpi=250)
```
