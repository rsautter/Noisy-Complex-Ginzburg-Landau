### Libraries

```python
from GPA import GPA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
ga = GPA(0.03)
ga.version()
```

# Without Noise

```python
data = np.load("data/withoutNoise/data.npy")
times = np.load("data/withoutNoise/times.npy")

print(data.shape)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(np.real(data[1000]),cmap='rainbow')
plt.title("Real(Z)",fontsize=15)
plt.subplot(1,3,2)
plt.imshow(np.imag(data[1000]),cmap='rainbow')
plt.title("Imag(Z)",fontsize=15)
plt.subplot(1,3,3)
plt.imshow(np.abs(data[1000]),cmap='rainbow')
plt.title("|Z|",fontsize=15)
plt.show()
```

```python
rz,iz,mz = [],[],[]
for i in tqdm(range(data.shape[0])):
    rz.append(ga(np.real(data[i]),moment=["G1","G2","G3"]))
    iz.append(ga(np.imag(data[i]),moment=["G1","G2","G3"]))
    mz.append(ga(np.abs(data[i]),moment=["G1","G2","G3"]))
rz,iz,mz = pd.DataFrame(rz),pd.DataFrame(iz),pd.DataFrame(mz)
```

```python
rz.to_csv("data/withoutNoise/gpa/real.csv", index=False)
iz.to_csv("data/withoutNoise/gpa/imag.csv", index=False)
mz.to_csv("data/withoutNoise/gpa/mod.csv", index=False)
```

```python
legendSize = 15
titleSize = 18

plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.title("Re(Z)",fontsize=titleSize)
#plt.plot(times, rz["G1"][:-1],color='b')
plt.plot(times, rz["G2"][:-1],color='r',label=r"$G_2$")
plt.plot(times, rz["G3"][:-1],color='k',label=r"$G_3$")
plt.legend(fontsize = legendSize)
plt.xticks([])
plt.xlim(0,max(times))

plt.subplot(3,1,2)
plt.title("Im(Z)",fontsize=titleSize)
#plt.plot(times, iz["G1"][:-1],color='b')
plt.plot(times, iz["G2"][:-1],color='r')
plt.plot(times, iz["G3"][:-1],color='k')
plt.xticks([])
plt.xlim(0,max(times))

plt.subplot(3,1,3)
plt.title("||Z||",fontsize=titleSize)
#plt.plot(times, mz["G1"][:-1],color='b')
plt.plot(times, mz["G2"][:-1],color='r')
plt.plot(times, mz["G3"][:-1],color='k')
plt.xlim(0,max(times))
plt.xlabel("Time")

plt.show()
```

# Real Part Comparison

```python
diff = pd.read_csv("data/diffusiveNoise/gpa/real.csv")
tdiff = np.load("data/diffusiveNoise/times.npy")

add = pd.read_csv("data/additiveNoise/gpa/real.csv")
tadd = np.load("data/additiveNoise/times.npy")

mul = pd.read_csv("data/multiplicativeNoise/gpa/real.csv")
tmul= np.load("data/multiplicativeNoise/times.npy")

without = pd.read_csv("data/withoutNoise/gpa/real.csv")
twithout = np.load("data/withoutNoise/times.npy")
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,1,1)
plt.plot(twithout, without["G2"][:-1],label="Without Noise", ls= '--')
plt.plot(tadd, add["G2"][:-1],label="Additive Noise", ls= '--')
plt.plot(tmul, mul["G2"][:-1],label="Multiplicative Noise", ls= '--')
plt.plot(tdiff, diff["G2"][:-1],label="Diffusive Noise", ls= '--')
plt.legend()
plt.title("$G_2\Re (Z)$", fontsize=20)
plt.xlim(0,max(twithout))

plt.subplot(2,1,2)
plt.plot(twithout, without["G3"][:-1],label="Without Noise", ls= '--')
plt.plot(tadd, add["G3"][:-1],label="Additive Noise", ls= '--')
plt.plot(tmul, mul["G3"][:-1],label="Multiplicative Noise", ls= '--')
plt.plot(tdiff, diff["G3"][:-1],label="Diffusive Noise", ls= '--')
plt.legend()
plt.title("$G_3(\Re (Z))$", fontsize=20)
plt.xlim(0,max(twithout))

plt.show()
```

# Delay Analysis

```python
rdiff = pd.read_csv("data/diffusiveNoise/gpa/real.csv")
idiff = pd.read_csv("data/diffusiveNoise/gpa/imag.csv")

radd = pd.read_csv("data/additiveNoise/gpa/real.csv")
iadd = pd.read_csv("data/additiveNoise/gpa/imag.csv")

rmul = pd.read_csv("data/multiplicativeNoise/gpa/real.csv")
imul = pd.read_csv("data/multiplicativeNoise/gpa/imag.csv")

rwithout = pd.read_csv("data/withoutNoise/gpa/real.csv")
iwithout = pd.read_csv("data/withoutNoise/gpa/imag.csv")
```

```python
labelSize=15
titlesize = 20

plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.title("Without Noise",fontsize=titlesize,weight="bold")
plt.plot(rwithout['G2'],iwithout['G2'],ls='-.',color='blue')
plt.xlabel("$G_2(\Re (z))$",fontsize=labelSize)
plt.ylabel("$G_2(\Im (z))$",fontsize=labelSize)

plt.subplot(2,2,2)
plt.title("Additive Noise",fontsize=titlesize,weight="bold")
plt.plot(radd['G2'],iadd['G2'],ls='-.',color='orange')
plt.xlabel("$G_2(\Re (z))$",fontsize=labelSize)
plt.ylabel("$G_2(\Im (z))$",fontsize=labelSize)

plt.subplot(2,2,3)
plt.title("Multiplicative Noise",fontsize=titlesize,weight="bold")
plt.plot(rmul['G2'],imul['G2'],ls='-.',color='green')
plt.xlabel("$G_2(\Re (z))$",fontsize=labelSize)
plt.ylabel("$G_2(\Im (z))$",fontsize=labelSize)

plt.subplot(2,2,4)
plt.plot(rdiff['G2'],idiff['G2'],ls='-.',color='red')
plt.title("Diffusive Noise",fontsize=titlesize,weight="bold")
plt.xlabel("$G_2(\Re (z))$",fontsize=labelSize)
plt.ylabel("$G_2(\Im (z))$",fontsize=labelSize)
plt.show()
```
