```python
from NCGL import NCGL
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
gl = NCGL(c1=0.0, c2=1.5, sigma_r=0.0,noiseArgs={"beta":2},msize=32,dim=3)
data,times = gl.solveRKF45(dt=0.03,ntimes=2500,stepsave=np.arange(4500))
```

```python
import numpy as np
import plotly.graph_objects as go

l = 32
it = 2090
X, Y, Z = np.mgrid[:l, :l, :l]
vol = np.real(data[it])

title = "Iteration: "+str(it)+", Time: "+str(np.round(times[it],2))
fig = go.Figure(data=go.Volume(
    x=X.flatten(), y=Y.flatten(), z=Z.flatten(),
    value=vol.flatten(),
    isomin=0.05,
    isomax=0.9,
    opacity=0.1,
    surface_count=25,
    colorscale='Rainbow'
    ))
fig.update_layout(width=700, height=700,title = title)

fig.show()
```

```python
with open('snaps.pkl', 'wb') as file: 
    pkl.dump(data, "snaps.pkl")
```
